# 1. usa.gov data from bit.ly

* Common form of webd data -> JSON. For example



```python
path = 'usagov_bitly_data2012-03-16-1331923249.txt'
open(path).readline()
```

* List Comprehension -> which enables the operations like json.loads


```python
import json
records = [json.loads(line) for line in open(path)]
records[0]
```

It's now easy to access

```python
records[0]['tz']
```

# Counting Time Zones in Pure Python

* find the most often-occuring time zones in the data set
* if condition implemented beacuse not all the records include 'tz' key


```python
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
```

* Now to produce counts by time zones 
* Initially using python standard library

```python
from collections import defaultdict

def get_counts2(seq):
    counts = defaultdict(int)
    for x in seq:
        counts[x] +=1
    return counts

counts = get_counts2(time_zones)
counts['America/New_York']
```

```python
len(time_zones)
```

* to get top 10 time zones

```python
def top_counts(count_dict, n=10):
    value_key_pairs = [(count,tz) for tz,count in count_dict.items()]
    value_key_pairs.sort(reverse=True)
    return value_key_pairs[:n]

top_counts(counts)
```

* Another way **collections.Counter** class, which makes this task a lot easier

```python
from collections import Counter
counts=Counter(time_zones)
counts.most_common(10)
```

# Counting Time Zones with pandas

* The main pandas data structure is the **DataFrame**, --> representing a table of data

```python
from pandas import DataFrame, Series
import pandas as pd

frame = DataFrame(records)
frame
```

* Output shown in summary view, shown by DataFrame Objects.
* counting the time_zones with DataFrame

```python
tz_counts = frame['tz'].value_counts()
tz_counts[:10]
```

* **fillna Function** can replace missing (NA) values and unknown(empty srings) values can be replaced by boolean array indexing:

```python
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz==''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10]
```

* a horizontal plot can be accomplished using the **plot** method on the counts

```python
%matplotlib inline
tz_counts[:10].plot(kind='barh', rot=0)
```

```python
results = Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
results.value_counts()[:8].plot(kind='barh', rot=0)
```

* Now want to decompose the top time zones into Windows and non-Windows users

```python
import numpy as np
cframe = frame[frame.a.notnull()]
operating_system = np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
operating_system[:5]
```

* **groupby Functions** can be computed using **size**
* result is then reshaped to table with **unstack fucntion**

```python
by_tz_os = cframe.groupby(['tz', operating_system])
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

* Let's select the top overall time zones. To do so, construct an indirect inde array from the row counts in agg_counts

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

* use **take function** to select the rows in that order, then slice off the last 10 rows: 

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh', stacked=True)
```

```python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True)
```
